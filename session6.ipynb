This notebook defines functions for sin(x) and cos(x).

```
import math

def my_sin(x):
    return math.sin(x)

def my_cos(x):
    return math.cos(x)
```

```
n = 1000
start = 0
end = 2 * math.pi
step = (end - start) / n

x = start
for i in range(n):
    print(x, my_sin(x), my_cos(x))
    x += step
```
